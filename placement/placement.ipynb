```python
def gender(x):
    if x=='M':
        return 1
    else:
        return 2


def ssc_b(x):
    if x=='Central':
        return 1
    else:
        return 2


def hsc_b(x):
    if x=='Central':
        return 1
    else:
        return 2

def hsc_s(x):
    if x=='Science':
        return 1
    elif x=='Commerce':
        return 2
    else:
        return 3

def degree_t(x):
    if x=='Sci&Tech':
        return 1
    elif x=='Comm&Mgmt':
        return 2
    else:
        return 3

def workex(x):
    if x=='Yes':
        return 1
    else:
        return 0
    
def specialisation(x):
    if x=='Mkt&HR':
        return 1
    else:
        return 2
```

```python
final['workex'] = final['workex'].apply(workex)
```

```python
final['specialisation'] = final['specialisation'].apply(specialisation)
```

```python
def status(x):
    if x=='Placed':
        return 1
    else:
        return 0
```
